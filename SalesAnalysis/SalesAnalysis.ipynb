# Sales Analysis 

#### Importing libraries 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

#### Merging all data files into single file.

```python
files = [file for file in os.listdir('./Sales_Data')]

all_months_data = pd.DataFrame()

for file in files :
    df = pd.read_csv("./Sales_Data/"+file)
    all_months_data = pd.concat([all_months_data, df])
all_months_data.to_csv("all_data.csv", index=False)
```

#### Reading all months data in one file

```python
all_data = pd.read_csv("all_data.csv")
all_data.head()
```

### Clean up the data

#### Drop Null ROWS

```python
nan_df = all_data[all_data.isna().any(axis = 1)]
nan_df.head()
all_data = all_data.dropna(how = 'all')
all_data.head()
```

#### delete Duplicate data 

```python
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
```

### Convert into correct data type

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])

all_data.head()
```

#### Add Month Column

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype(np.int32)
all_data.head()
```

### Add Sales Column 

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data['Sales'] = pd.to_numeric(all_data['Sales'])
all_data.head()
```

### Add City Column

```python
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x : f"{get_city(x)} ({get_state(x)})")
all_data.head()
```

# Data Exploration 

#### What are the best month for sales ?  and how much was earned that month ?

```python
result = all_data.groupby('Month')[['Quantity Ordered', 'Price Each','Sales']].sum()
```

```python
months = range(1,13)

plt.figure(figsize=(9,6))
plt.bar(months, result['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.ticklabel_format(style='plain', axis='y')

plt.xlabel('Month Number')
plt.show()
```

#### December recorded the highest sales with a total revenue of approximately $4.6M, making it the strongest month of the year.

#### What City had th highest number of sales ?

```python
result = all_data.groupby('City')[['Quantity Ordered', 'Price Each','Sales']].sum()
result
```

```python
cities = [city for city, df in all_data.groupby('City')]

plt.figure(figsize=(9,6))
plt.bar(cities, result['Sales'])
plt.xticks(cities, rotation = 'vertical')
plt.ylabel('Sales in USD ($)')
plt.ticklabel_format(style='plain', axis='y')

plt.xlabel('City')
plt.show()
```

#### San Francisco generated the highest sales among all cities, with revenue exceeding $8.2M.
#### This indicates strong demand and purchasing power in this region, making it a key market for the business.

#### What time should we display advertisements to maximize likelihood of customer's buying product?

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
```

```python
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data['Minute'] = all_data['Order Date'].dt.minute
all_data.head()
```

```python
hours = [hour for hour, df in all_data.groupby('Hour')]

plt.plot(hours, all_data.groupby(['Hour']).count())
plt.xticks(hours)
plt.xlabel('Hour')
plt.ylabel('Number of Orders')
plt.grid()
plt.show()
         
```

#### The highest customer activity occurs around 78 PM. Therefore, displaying advertisements during 68 PM will maximize visibility and conversion, with 11 AM1 PM serving as a strong secondary window.

#### What products are most often sold together ? 

```python
df = all_data[all_data['Order ID'].duplicated(keep = False)].copy()
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x : ','.join(x))
df = df[['Order ID','Grouped']].drop_duplicates()
df.head()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
for key, value in count.most_common(10):
    print(key, value)
```
